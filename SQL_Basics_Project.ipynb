```python
%%sql
SELECT MIN(population), MAX(population), MIN(population_growth), MAX(population_growth)
FROM facts
```

```python
%%sql
SELECT *
FROM facts
WHERE population = 0
```

```python
%%sql
SELECT *
FROM facts
WHERE population = 7256490011	
```

```python
%%sql
SELECT MIN(population), MAX(population), MIN(population_growth), MAX(population_growth)
FROM facts
WHERE name <> 'World'
```

```python
%%sql
SELECT AVG(population), AVG(area)
FROM facts
```

```python
%%sql
SELECT *
FROM facts
WHERE population > (SELECT AVG(population) FROM facts WHERE name <> 'World') AND area < (SELECT AVG(area) FROM facts WHERE name <> 'World') 
```

```python
%%sql
SELECT name, population
FROM facts
WHERE name <> 'World'
ORDER BY population desc
LIMIT 1
```

From the data we can see that China has the highest population.

```python
%%sql
SELECT name, ROUND(CAST(area_water AS FLOAT)/area_land, 2) AS ratio
FROM facts
ORDER BY ratio desc
```

The country/territory with the highest ratio of water to land is the Virgin Islands.
